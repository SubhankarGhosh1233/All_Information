```python
class dun:
  def __init__(self,x,y):
    self.x = x
    self.y =y
  def __add__(self,other):
    return  (self.x +other.x,self.y + other.y)

  def __getitem__(self,index):
    if index==0:
      return self.x
    elif index ==1:
      return self.y
    else:
      return "out of range"
```

```python
p = dun(2,3)
q = dun(4,3)
# p+q
```

```python
L1 = [1, 2, 3]
L2 = [1, 2, 3, 4, 5]

set1 = set(L1)
set2 = set(L2)

is_subset = set1.issubset(set2)

if is_subset:
    print("L1 is a subset of L2")
else:
    print("L1 is not a subset of L2")
```

```python
l1 = [1,2,50]
l2 = [1,2,3,4,5]

for i in l1:
    if i not in l2:
        print("Not a subset")
        break

    else:
        print("It is a subset")
```
